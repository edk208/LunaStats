```python
from terra_sdk.client.lcd import LCDClient
from terra_sdk.core.tx import Tx
import time
```

```python
terra = LCDClient(chain_id="columbus-5", url="https://lcd.terra.dev")
```

```python
#Current block
terra.tendermint.block_info()['block']['header']['height']
```

```python
#Number of transactions in current block
block = terra.tendermint.block_info()
print(len(block['block']['data']['txs']))
```

```python
# Count number of Transactions and WASM dApp transactions.  Ignore Oracle broadcast
count_tx = 0
count_dapps = 0
for b in range(len(block['block']['data']['txs'])):
    tx = block['block']['data']['txs'][b]
    tx = terra.tx.decode(tx)
    if (tx.body.to_data()['messages'][0].get('@type')) == '/cosmos.bank.v1beta1.MsgSend':
        count_tx = count_tx+1
    if (tx.body.to_data()['messages'][0].get('@type')) == '/terra.wasm.v1beta1.MsgExecuteContract':
        count_dapps = count_dapps+1
    #print((tx.body.to_data()['messages'][0].get('@type')))
print('Count of TXs: ',count_tx)
print('Count of dApps: ',count_dapps)
```

```python
#Luna has an average block time around 6 seconds
#Number of blocks per minute around 10, or 600 blocks per hour
#Sample the past 6 hours or 3600 blocks, 7667505 -> 7671105

total_tx = 0
total_msgsend = 0
total_dapps = 0

for bnum in range(7667505,7671105):
    print('Processing Block:', bnum)
    block = terra.tendermint.block_info(bnum)
    total_tx = total_tx + len(block['block']['data']['txs'])
    count_tx = 0
    count_dapps = 0
    for b in range(len(block['block']['data']['txs'])):
        tx = block['block']['data']['txs'][b]
        try: 
            tx = terra.tx.decode(tx)
        except TypeError:
            pass
        except AttributeError:
            pass
        else:
            if (tx.body.to_data()['messages'][0].get('@type')) == '/cosmos.bank.v1beta1.MsgSend':
                count_tx = count_tx+1
            if (tx.body.to_data()['messages'][0].get('@type')) == '/terra.wasm.v1beta1.MsgExecuteContract':
                count_dapps = count_dapps+1
    total_msgsend = total_msgsend + count_tx
    total_dapps = total_dapps + count_dapps
    #Don't Spam the Terra Endpoint API too hard
    time.sleep(0.01)
    
print('Count of TXs: ',total_tx)    
print('Count of MSG TXs: ',total_msgsend)
print('Count of dApps TXs: ',total_dapps)
```
